Tentando montar uma função:

```python
def RIMAS(arquivo, d, n):
  dados = pd.read_csv(
    arquivo,
    encoding='latin-1',
    delimiter=';',
    skiprows=3,
    usecols=[1, 2, 3],
    decimal=",",
)
  dados.dropna(axis=0, inplace=True)
  dados["data"] = pd.to_datetime(dados["Data da medição"] + dados["Hora da medição"], format="%d/%m/%Y%H:%M")
  dados["tempo_s"] = (dados.data - pd.to_datetime("2000-01-01")).dt.total_seconds()
  dados.rename(columns={'Nível da água': 'niveldagua_m', 'Data da medição' : 'data_da_medicao', 'Hora da medição': 'hora_da_medicao'}, inplace=True)

  datetime.datetime.today().isoformat()
  AQ = arquivo.split("/")[-4]
  Form = arquivo.split("/")[-3]
  title = arquivo.split("/")[-2]

  with open(title + "PND.csv", "w") as f:
    f.write(f"# Processado em: {datetime.datetime.today().isoformat()}\n")
    dados.to_csv(f)

  caminho = Path("/content/drive/My Drive/IC/BaseDados/RIMAS/" + AQ + "/" + Form + "/" + title)
  shutil.move(title + "PND.csv", caminho)

  if d == 1:
    plt.figure(figsize=(15, 6))
    plt.gca().invert_yaxis()  # Inverter o eixo y para que o maior valor esteja no topo
    plt.title(title + " Data", fontsize= 25)
    plt.xlabel('Data', fontsize= 15)
    plt.ylabel('Nível da água (m)', fontsize= 15)
    plt.scatter(dados['data'], dados["niveldagua_m"], color= 'blue', marker= 'o', s= 5)
    plt.tick_params(labelsize= 10)
    plt.tight_layout()
    plt.grid()
    plt.savefig("Data.png")
    shutil.move("Data.png", caminho)

  if n == 1:
    plt.figure(figsize=(15, 6))
    plt.gca().invert_yaxis()
    plt.title(title + " Tempo númerico", fontsize= 25)
    plt.xlabel('Tempo (s)', fontsize= 15)
    plt.ylabel('Nível da água (m)', fontsize= 15)
    plt.scatter(dados.tempo_s, dados["niveldagua_m"], color= 'blue', marker= 'o', s= 5)
    plt.tick_params(labelsize= 10)
    plt.tight_layout()
    plt.grid()
    plt.savefig("TempoNumerico.png")
    shutil.move("TempoNumerico.png", caminho)

  return title + "PND.csv"
```
